```python
# visualization
import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
time_multi_round = [10.42, 10.45, 10.50]

time_seq = [6.5, 10, 12.5]

sns.set_style("white")


plt.bar([1,2,3],time_multi_round, alpha=0.4, label='multi round', color='blue')
plt.bar([1,2,3], time_seq, alpha=0.4, label = 'sequential', color='green')

plt.ylabel('time in hours')

plt.xticks([1,2,3], ('step1','step2','step3'))

plt.legend()
```
