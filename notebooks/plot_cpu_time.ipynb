```python
# visualization
import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
time_multi_round = [10.8, 10.8, 10.8, 32.4]

time_seq = [4.2, 6.73, 11.22, 22.15]

sns.set_style("white")


plt.bar([1,2,3,4],time_multi_round, alpha=0.4, label='multi round SNPE', color='blue')
plt.bar([1,2,3,4], time_seq, alpha=0.4, label = 'incremental approach', color='green')

plt.ylabel('time in hours')

plt.xticks([1,2,3, 4], ('step1','step2','step3', 'total'))

plt.title('CPU hours')

plt.legend()
```
