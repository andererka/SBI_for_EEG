# Fig. 4.7 for Thesis: Time comparison of SNPE and NIPE

```python
# visualization
import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns
```

```python


time_multi_round = [12.5, 13, 14, 39.5]

time_seq = [5, 9.75, 12.75, 27.5]

time_seq2 = [0.5, 7.25, 20, 27.75]

sns.set_style("white")


plt.bar([1,3,5,7],time_multi_round, alpha=0.6, label='SNPE', color='green')
plt.bar([1.3,3.3,5.3,7.3], time_seq, alpha=0.8, label = 'NIPE no ratio', color='blue')

plt.bar([1.6,3.6,5.6,7.6], time_seq2, alpha=0.8, label = 'NIPE ratio', color='purple')

plt.ylabel('time in hours')

plt.xticks([1,3,5, 7], ('step1','step2','step3', 'total'))

#plt.title('CPU hours')

sns.despine()

plt.legend()

plt.show()

```
